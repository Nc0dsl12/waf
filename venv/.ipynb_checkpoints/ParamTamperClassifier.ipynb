```python
X = df['length'].to_numpy().astype(str)
y = df['label'].to_numpy().astype(str)
```

```python
print(len(X))
print(len(y))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
cross_val_score(clf, X.reshape(-1, 1), y, cv=10)
```

```python
clf.fit(X.reshape(-1, 1), y)
```

```python
import pickle

# Save the model
with open('pt_predictor.pkl', 'wb') as f:
    pickle.dump(clf, f)

# To verify it worked, you can load it:
with open('pt_predictor.pkl', 'rb') as f:
    loaded_clf = pickle.load(f)
```
